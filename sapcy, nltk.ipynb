NLTK Stemming

Porter Stemmer

```python
# Stemming is the process of reducing a word to its root form. It is a common technique used in natural language processing (NLP) to 
# simplify words and reduce the number of unique words in a text. The Porter Stemmer is one of the most popular stemming algorithms, 
# which uses a set of rules to remove suffixes from words.

import nltk

from nltk.stem.porter import PorterStemmer
p_stemer=PorterStemmer()
words=["run","runner","running","ran","runs","easily","fairly"]
for word in words:
    print(f"{word} -> {p_stemer.stem(word)}")
```

Snowball stemmer

```python
# The Snowball Stemmer is an improved version of the Porter Stemmer, which uses a more comprehensive set of rules to stem words.
# It is also known as the "English Stemmer" and is available in the NLTK library.

from nltk.stem import SnowballStemmer
s_stemer=SnowballStemmer(language='english')
for word in words:
    print(f"{word} -> {s_stemer.stem(word)}")
```

Lemmatization

```python
# Lemmatization is the process of reducing a word to its base form, called a lemma. 
# It is similar to stemming but uses a more sophisticated approach that takes into account the context and meaning of the word. 
# The spaCy library provides a powerful lemmatization tool that can be used to lemmatize words in a text.

import spacy
nlp=spacy.load('en_core_web_sm')
def lem(text):
    for token in nlp(text):
        print(f"{token.text} -> {token.lemma_} -> {token.pos_}")
lem(u"run runner running ran runs easily fairly")
```

Viewing Matches

```python
from spacy.matcher import Matcher
matcher=Matcher(nlp.vocab)
```

```python
pattern1=[{'LOWER':'solarpower'}]
pattern2=[{'LOWER':'solar'},{'LOWER':'power'}]
pattern3=[{'LOWER':'solar'},{'IS_PUNCT':True},{'LOWER':'power'}]
pattern4=[{'LOWER':'solarpower'},{'IS_PUNCT':True},{'LOWER':'power'}]
pattern5=[{'LOWER':'solar'},{'LOWER':'power'}]
matcher.add("SolarPower",[pattern1,pattern2,pattern3,pattern4,pattern5])
```

```python
doc=nlp(u"The solar Power industry continues to grow as demand \
        for solarpower increases. Solar-power cars are gaining popularity.")
matches=matcher(doc)
for match_id, start, end in matches:
    string_id=nlp.vocab.strings[match_id]
    span=doc[start:end]
    print(match_id,string_id,start,end,span.text)
```

```python
import spacy
from spacy.matcher import PhraseMatcher

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Define phrase list
phrase_list = [
    'voodoo economics',
    'supply-side economics',
    'Keynesian economics',
    'monetarism',
    'fiscal stimulus',
    'trickle-down theory',
    'deficit spending',
    'economic liberalization',
    'market equilibrium',
    'inflation targeting'
]

# Convert phrases into spaCy Doc objects
patterns = [nlp.make_doc(text) for text in phrase_list]

# Initialize PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
matcher.add("ECON_PHRASES", patterns)

# Economics text
text = """
Economic debates have often revolved around contrasting schools of thought. 
During the Reagan era, critics labeled certain policies as 'voodoo economics', 
while supporters defended them as 'supply-side economics'. In contrast, 
'Keynesian economics' argues for active government intervention, especially 
through 'fiscal stimulus' and 'deficit spending' during recessions.

Meanwhile, 'monetarism' stresses the importance of controlling the money supply 
to manage inflation, often linked with 'inflation targeting' by central banks. 
Some economists advocate the 'trickle-down theory', suggesting that benefits 
for the wealthy eventually reach lower-income groups. 

In developing nations, 'economic liberalization' policies have aimed to open 
markets, encourage foreign investment, and achieve 'market equilibrium'.
"""

# Process text
doc = nlp(text)

# Apply matcher
matches = matcher(doc)

# Print results
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get rule name
    span = doc[start:end]                    # Matched span
    print(match_id, string_id, start, end, span.text)
```
