Basic Level

```python
#Write a regex to extract all numbers from
import re
text="My roll number is 42 and age is 25"
re.findall(r'\d+',text)
```

```python
#Find all words starting with a
text="apple banana apricot mango"
re.findall(r'\ba\w+',text)
```

```python
# 3. Check if a string contains only alphabets
text1, text2 = "HelloWorld", "Hello123"
print(bool(re.findall(r'^[A-Za-z]+$', text1)))
print(bool(re.findall(r'^[A-Za-z]+$', text2)))
```

```python
# 4. Extract email addresses
text = "Contact me at test123@gmail.com or admin@company.org"
re.findall(r'[\w\.-]+@[\w\.-]+', text)
```

```python
# 5. Split a sentence into words
text = "Regex is powerful!"
print(re.split(r'\W+', text))   # Explanation: \W+ splits on non-word characters.
print(re.split(r'\s+', text))   # Explanation: \s+ splits on whitespace characters.
```

Intermediate Level

```python
# 1. Validate phone numbers
text1, text2 = "9876543210", "98765abc10"
print(bool(re.findall(r'^[6-9]\d{9}$', text1))) # Explanation: Starts with 6–9, followed by 9 digits.
print(bool(re.findall(r'^[6-9]\d{9}$', text2)))
```

```python
# 2. Extract hashtags
text = "Loving #Python and #Regex"
re.findall(r'#\w+', text)
```

```python
# 3. Find repeated words
text = "This is is a test test"
re.findall(r'\b(\w+)\s+\1\b', text)
```

NLTK (Natural Language Tool kit)

```python
import nltk
from nltk.stem import PorterStemmer

ps=PorterStemmer()
words=['run','runner','running','ran','runs','easily','fairly']
for word in words:
    print(word+' --> '+ps.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
snow_stemmer=SnowballStemmer("english")
for word in words:  
    print(word+' --> '+snow_stemmer.stem(word))
```

```python
from nltk.stem.snowball import SnowballStemmer
snow_stemmer=SnowballStemmer("english")
words=['genrous','generously','generation','generational']
for word in words:  
    print(word+' --> '+snow_stemmer.stem(word))
```

Lemmatization

```python
import spacy
nlp=spacy.load("en_core_web_sm")
doc=nlp("I am a runner running in a race because I love to run since I ran today")
for token in doc:
    print(token.text,'\t',token.pos_,'\t',token.lemma_)
```

```python
def show_lemmas(text):
    for token in text:
        print(f"{token.text:12} {token.pos_:10} {token.lemma_}")    
doc1=nlp(u"I saw eighteen mice today!")
show_lemmas(doc1)
```
